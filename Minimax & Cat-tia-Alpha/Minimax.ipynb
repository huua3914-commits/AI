```python
import copy, math, random, numpy

X = "X"
O = "O"
EMPTY = None
user = None
ai = None

def initial_state():
    """ Returns starting state of the board. """
    return [[EMPTY, EMPTY, EMPTY], [EMPTY, EMPTY, EMPTY], [EMPTY, EMPTY, EMPTY]]

def player(board):
    """ Returns player who has the next turn on a board. """
    count = 0
    for i in board:
        for j in i:
            if j:
                count += 1
    if count % 2 != 0:
        return ai
    return user

def actions(board):
    """ Returns set of all possible actions (i, j) available on the board. """
    res = set()
    for i in range(3):
        for j in range(3):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    """ Returns the board that results from making move (i, j) on the board. """
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board

def winner(board):
    """ Returns the winner of the game, if there is one. """
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != EMPTY: return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != EMPTY: return board[0][i]
    if board[0][0] == board[1][1] == board[2][2] != EMPTY: return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != EMPTY: return board[0][2]
    return None

def terminal(board):
    """ Returns True if game is over, False otherwise. """
    if winner(board) is not None: return True
    return all(cell != EMPTY for row in board for cell in row)

def utility(board):
    """ Returns 1 if X won, -1 if O won, 0 otherwise. """
    res = winner(board)
    if res == X: return 1
    if res == O: return -1
    return 0

def maxValue(state):
    if terminal(state): return utility(state)
    v = -math.inf
    for action in actions(state): v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state): return utility(state)
    v = math.inf
    for action in actions(state): v = min(v, maxValue(result(state, action)))
    return v

def minimax(board):
    curr_player = player(board)
    best_move = None
    if curr_player == X:
        v = -math.inf
        for action in actions(board):
            m = minValue(result(board, action))
            if m > v:
                v = m
                best_move = action
    else:
        v = math.inf
        for action in actions(board):
            m = maxValue(result(board, action))
            if m < v:
                v = m
                best_move = action
    return best_move

if __name__ == "__main__":
    board = initial_state()
    print("Choose your side: X or O")
    user = input().upper()
    ai = O if user == X else X

    while not terminal(board):
        curr_player = player(board)
        if curr_player == user:
            print("Your turn. Enter row,col (0-2):")
            while True:
                try:
                    r = int(input("Row: "))
                    c = int(input("Col: "))
                    if board[r][c] == EMPTY:
                        board = result(board, (r, c))
                        break
                    else: print("Already taken.")
                except (ValueError, IndexError): print("Invalid input.")
        else:
            print("AI is thinking...")
            board = result(board, minimax(board))
        print(numpy.array(board))

    res = winner(board)
    print("Game Over: " + (f"{res} wins!" if res else "Tie."))
```
